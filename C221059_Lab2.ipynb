```python
import cv2
import matplotlib.pyplot as plt

def show_image(title, img, cmap=None):

    if len(img.shape) == 3:
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(img, cmap=cmap)
    plt.title(title)
    plt.axis('off')
    plt.show()

image = cv2.imread("rgb.jpg")
if image is None:
    print("Error: Could not read image.")
    exit()


gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
show_image("Grayscale Image", gray_image, cmap='gray')


_, binary_image = cv2.threshold(gray_image, 128, 255, cv2.THRESH_BINARY)
show_image("Binary Image", binary_image, cmap='gray')


def reduce_levels(img, levels):

    factor = 256 // levels
    reduced_img = (img // factor) * factor
    return reduced_img

for levels in [64, 32, 16]:
    reduced_img = reduce_levels(gray_image, levels)
    show_image(f"Reduced Levels - {levels}", reduced_img, cmap='gray')
```
