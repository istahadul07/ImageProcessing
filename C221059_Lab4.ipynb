```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the uploaded image (grayscale)
img_path = '/content/df.png'
img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)

# Threshold parameters
t1, t2 = 100, 150

# Single Thresholding
_, single_thresh = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)

# Dual Thresholding (between two thresholds)
dual_thresh = cv2.inRange(img, t1, t2)

# Gray Level Slicing 1st approach
def gray_level_slicing_1st(image, t1, t2):
    output = np.copy(image)
    mask = (image >= t1) & (image <= t2)
    output[mask] = 255
    return output

# Gray Level Slicing 2nd approach
def gray_level_slicing_2nd(image, t1, t2):
    output = np.zeros_like(image)
    mask = (image >= t1) & (image <= t2)
    output[mask] = 255
    return output

gls1 = gray_level_slicing_1st(img, t1, t2)
gls2 = gray_level_slicing_2nd(img, t1, t2)

# Improved plot with larger figures and arranged in two rows
fig, axs = plt.subplots(2, 3, figsize=(18, 10))

axs[0,0].imshow(img, cmap='gray')
axs[0,0].set_title('Original Grayscale')
axs[0,0].axis('off')

axs[0,1].imshow(single_thresh, cmap='gray')
axs[0,1].set_title('Single Threshold (127)')
axs[0,1].axis('off')

axs[0,2].imshow(dual_thresh, cmap='gray')
axs[0,2].set_title('Dual Threshold (100-150)')
axs[0,2].axis('off')

axs[1,0].imshow(gls1, cmap='gray')
axs[1,0].set_title('Gray Level Slicing\n1st Approach')
axs[1,0].axis('off')

axs[1,1].imshow(gls2, cmap='gray')
axs[1,1].set_title('Gray Level Slicing\n2nd Approach')
axs[1,1].axis('off')

# Hide the last empty subplot
axs[1,2].axis('off')

plt.tight_layout()
plt.show()
```
