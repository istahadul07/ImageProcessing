```python
import cv2
import matplotlib.pyplot as plt

def show_image(title, img):

    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(img_rgb)
    plt.title(title)
    plt.axis('off')
    plt.show()


image = cv2.imread("c221059.jpg")
if image is None:
    print("Error: Could not read image.")
    exit()


show_image("Original Image", image)


bright_image = cv2.convertScaleAbs(image, alpha=1, beta=70)  # Increase beta to brighten


show_image("Brightened Image", bright_image)


dark_image = cv2.convertScaleAbs(image, alpha=1, beta=-50)  # Decrease beta to darken


show_image("Darkened Image", dark_image)
```
